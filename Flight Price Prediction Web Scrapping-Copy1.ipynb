```python
urls = []

source = ["BLR-Bengaluru","DEL-Delhi"]
desti = ["CCU-Kolkata","BOM-Mumbai","MAA-Chennai","HYD-Hyderabad"]

# passing random date on which flight is schedule:

dates = ["2022-11-03","2022-11-04","2022-12-05","2022-11-08","2022-12-03","2022-11-15","2022-11-24","2022-11-28","2022-12-09","2022-12-15"]

for a in range(0,2):
    for b in range(0,4):
            for c in range(0,10):
                url="https://paytm.com/flights/flightSearch/"+source[a]+"/"+desti[b]+"/1/0/0/E/"+dates[c] #as url is a string we can manipulate it and make changes according to our needs
                urls.append(url)
```

```python
len(urls)
```

```python
# making list of features i am going to fetch 
name=[]
departure_time=[]
arrival_time=[]
duration=[]
price=[]
stop=[]
place=[]
date=[]
for url in urls:
    driver.get(url)
    time.sleep(5)
    
    for x in driver.find_elements(By.XPATH, '//div[@class="_3H-S _1Eia"]'):
        name.append(x.text)
            
    for y in driver.find_elements(By.XPATH, '//div[@class="_3H-S"]'):
        departure_time.append(y.text)
        
    for d in driver.find_elements(By.XPATH, '//div[@class="_3H-S _1wD5"]'):
        arrival_time.append(d.text.split("\n")[0])
        
    for e in driver.find_elements(By.XPATH, '//div[@class="vY4t"]'):
        duration.append(e.text)
        
    for f in driver.find_elements(By.XPATH, '//div[@class="_1cxG"]'):
        price.append(int(f.text.replace(",","")))
        
    for g in driver.find_elements(By.XPATH, '//div[@class="_7BOG"]'):
        stop.append(g.text.split()[0])
        
    for h in driver.find_elements(By.XPATH, '//div[@class="NqXj"]'):
        place.append(h.text)
        
    for i in range(0,len(driver.find_elements(By.XPATH, '//div[@class="_1cxG"]'))):
               date.append(url.split('/')[11])
```

```python
source=place[0::2]
destination=place[1::2]
```

```python
print(len(arrival_time),len(duration),len(price),len(name),len(departure_time),len(source),len(destination),len(date))
```

```python
# this is the dataframe i have fetched
flights=pd.DataFrame({"Date":date,"Flight name":name,"Source":source,"Destination":destination,"Departure_time":departure_time,
                    "Arrival_time":arrival_time,"Duration":duration,"No of stop":stop,"Price":price})
                 
```
