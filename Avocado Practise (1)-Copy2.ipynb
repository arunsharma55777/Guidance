## Model Building :

```python
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.ensemble import GradientBoostingRegressor, BaggingRegressor
from xgboost import XGBRegressor
```

```python
# scaling the datset:

sc = StandardScaler()
x_scaled = sc.fit_transform(x)
```

## Training the model

### Finding the best random state : 

```python
lr = LinearRegression()

for i in range(0,100):
    x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size =0.20, random_state =i)
    lr.fit(x_train, y_train)
    pred_train = lr.predict(x_train)
    pred_test = lr.predict(x_test)
    print(f"At random state {i}, the training accuracy is : {r2_score(y_train, pred_train)}" )
    print(f"At random state {i}, the testing accuracy is : {r2_score(y_test, pred_test)}" )
    print('\n')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.20, random_state = 51)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import cross_val_score 
```

```python
x_train, cv_x, y_train, cv_y = train_test_split(x_train, y_train, test_size = 0.20)
```

```python
# checking the shape of x_train, x_test, y_train, y_test, cv_x, cv_y

x_train.shape, x_test.shape, y_train.shape, y_test.shape, cv_x.shape, cv_y.shape
```

```python
def fun(instance, x_train, y_train, x_test, y_test,cv_x, cv_y):
    instances = instance.fit(x_train,y_train)
    train_pred = instances.predict(x_train)
    test_pred = instances.predict(x_test)
    print('r2_score', r2_score(y_test, test_pred))
    print('Mean squared error:',mean_squared_error(y_test,test_pred))
    print('Root mean squared error:',np.sqrt(mean_squared_error(y_test, test_pred)))
    print('cross_val_score',cross_val_score(instances,cv_x,cv_y))
    print('cross_val_score Mean',cross_val_score(instances,cv_x,cv_y).mean())
```

### Linear Regression :

## Classification - Target - Region

```python
# Splitting the dataset into features(x) and target columns(y) :

X = df1.drop(columns = ['region'], axis =1)
Y = df1["region"]
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression,Ridge
from sklearn.ensemble import GradientBoostingClassifier, BaggingClassifier, RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
# scaling the datset:

sc = StandardScaler()
X_scaled = sc.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size = 0.20, random_state = 51)
```

```python
X_train, cv_X, Y_train, cv_Y = train_test_split(X_train, Y_train, test_size = 0.20)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape, cv_X.shape, cv_X.shape
```

```python
def func(instance, X_train, Y_train, X_test, Y_test, cv_X, cv_Y):
    instances = instance.fit(X_train,Y_train)
    train_pred = instances.predict(X_train)
    test_pred = instances.predict(X_test)
    print('training confucion matrix',confusion_matrix(Y_train,train_pred))
    print('testing confucion matrix',confusion_matrix(Y_test,test_pred))
    print('test ',classification_report(Y_test, test_pred))
    print('training ',classification_report(Y_train, train_pred))
    print('cross_val_score',cross_val_score(instances,cv_X,cv_Y))
    print('cross_val_score Mean',cross_val_score(instances,cv_X,cv_Y).mean())
```

```python
model1 = KNeighborsClassifier()
func(model1, X_train,Y_train,X_test,Y_test,cv_X,cv_Y)
```

```python
model2 = DecisionTreeClassifier()
func(model2, X_train,Y_train,X_test,Y_test,cv_X,cv_Y)
```

```python
model3 = GradientBoostingClassifier()
func(model3, X_train,Y_train,X_test,Y_test,cv_X,cv_Y)
```

```python
model4 = BaggingClassifier()
func(model4, X_train,Y_train,X_test,Y_test,cv_X,cv_Y)
```

```python
model5 = RandomForestClassifier()
func(model5, X_train,Y_train,X_test,Y_test,cv_X,cv_Y)
```

```python
model6 = SVC()
func(model6, X_train,Y_train,X_test,Y_test,cv_X,cv_Y)
```
